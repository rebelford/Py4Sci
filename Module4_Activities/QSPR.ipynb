````python
from rdkit import Chem
from mordred import Calculator, descriptors
import pandas as pd

# RDKit molecule
mol = Chem.MolFromSmiles("c1ccccc1O")  # phenol

# Set up Mordred calculator
calc = Calculator(descriptors, ignore_3D=True)

# Compute descriptors (Result object)
result = calc(mol)

# Zip descriptor names with values
descriptor_names = [str(d) for d in calc.descriptors]
descriptor_values = list(result)

# Create a DataFrame
df = pd.DataFrame({
    "Descriptor": descriptor_names,
    "Value": descriptor_values
})

# Preview first few rows
print(df.head())


# Show it in scrollable box
my_fun.scrollable_df(df, height='500px')
````
